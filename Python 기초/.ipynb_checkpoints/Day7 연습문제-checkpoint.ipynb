```python
#커피 자판기 만들기
#사용자 기능 : 동전 투입, 커피 선택, 잔액 반환/커피 선택
#관리자 기능 : 커피 재료, 동전 투입(1000,500,100,50,10), 품절 표시

#관리자 모드 기능
class admin:
    def add_ingrd(self,cof_add,cup_add,cof,cup): #재료 추가
        print('커피 %d잔 남아있습니다' %(cof+cof_add))
        print('컵이 %d개 남아있습니다' %(cup+cup_add))
    def add_change(self,add_ch,ch_now):
        res=[]
        for i in range(4):
            res.append(add_ch[i]+ch_now[i])
        print('1000원권 : %d개' %(res[0]))
        print(' 500원권 : %d개' %(res[1]))
        print(' 100원권 : %d개' %(res[2]))
        print(' 100원권 : %d개' %(res[3]))
        return res
        
#사용자 모드 기능
class user:
    def insert(self,add_mny,mny_now):
        res=mny_now
        amt=[1000,500,100,50]
        for j in range(4):
            res+=add_mny[j]*amt[j]
        print('잔액 : %d원' %(res))
        return res
    def select(self,num,cof,mny_now):
        res=(lambda x:x-200 if num==1 else x-250 if num==2 else x-300)(mny_now)
        if res<0:
            print('잔액을 초과하였습니다. 처음으로 돌아갑니다.')
            res=mny_now
        else:
            cof-=1
            print('커피가 나옵니다')
            print('잔액 : %d원' %res)
        return cof,res

    

cof=20                  #커피 20잔분으로 시작
cup=20                  #컵 20개로 시작
ch_now=[20,20,20,20]    #1000, 500, 100, 50 각 거스름돈 20개로 시작
mny_now=0               #시작 시 투입금 0


while  cof:
    prs1=int(input('1:관리자 모드\n2:사용자모드\n3.종료\n'))        #모드 선택
    if prs1==1:
        mode=admin()
        prs2=int(input('1:재료 투입\n2:거스름돈 투입\n3.종료\n'))   #관리자 모드에서 할 일 선택
        if prs2==1:                                                #관리자-재료 투입
            ad_ing,ad_cup=list(map(int,input('투입 재료 갯수, 투입 컵 갯수 :').split(',')))
            mode.add_ingrd(ad_ing,ad_cup,cof,cup)
        elif prs2==2:                                              #관리자-거스름돈 투입
            add_ch=list(map(int,input('추가할 1000원,500원,100원,50원권 갯수 :').split(',')))
            ch_now=mode.add_change(add_ch,ch_now)
        else:                                                      #관리자 모드 종료
            pass
        
    elif prs1==2:
        mode=user()
        prs3=int(input('1:돈 투입\n2:커피 선택\n3.잔액 반환\n'))   #사용자 모드에서 할 일 선택
        if prs3==1:                                               #사용자-돈 투입
            add_mny=list(map(int,input('투입 1000, 500, 100, 50원권 갯수 :').split(',')))
            mny_now=mode.insert(add_mny,mny_now)
        elif prs3==2:                                            #사용자-커피 선택
            if mny_now<200:
                print('잔액이 부족합니다')
                continue
            prs4=int(input('1:아메리카노\n2:믹스커피\n3:핫초코\n'))
            cof,mny_now=mode.select(prs4,cof,mny_now)
    else:
        break
```
