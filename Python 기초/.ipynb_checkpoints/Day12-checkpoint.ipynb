```python
# 검색어 입력하면 <투데이 에너지> 웹사이트에서 검색 결과(기사 제목 및 내용) 출력
# 아직 출력 형태가 깔끔하지 않아 전처리를 더 해야 함

import requests as req
from bs4 import BeautifulSoup as bs
from selenium import webdriver
import re

url='http://www.todayenergy.kr'
browser=webdriver.PhantomJS()
browser.get(url)

# 검색어 입력 후 검색 실행
sc_wd=browser.find_element_by_id('search')
sc_wd.clear()
sc_wd.send_keys(input('검색어를 입력해주세요 : '))
sc_wd.submit()

# 기사 리스트형으로 출력
list_bt=browser.find_element_by_css_selector('#user-container > div.float-center.max-width-1280 > div.user-content > section > article > div.article-list > header > div.float-right.button-group.small > a:nth-child(1)')
list_bt.click()

source=browser.page_source # 검색 결과에 대한 html 소스
url=browser.current_url
browser.quit()
soup=bs(source,'html.parser')

# 총 기사 건수 저장
num_dir='#user-container > div.float-center.max-width-1280 > div.user-content > section > article > div.article-list > header > div.float-left.article-list-header-title > small'
art_num=re.search('[\d,]+',soup.select_one(num_dir).string).group().split(',')
art_num=int(''.join(art_num))

page=0 # 몇 페이지의 기사인지 저장하는 변수
num=0  # n 페이지의 몇 번째 기사인지 저장하는 변수

for i in range(0,art_num,20):
    
    page+=1
    num=0  # n 페이지의 몇 번째 기사인지 저장하는 변수
    
    # 각 기사 url 주소 가져오기
    arti=soup.select('div.list-titles a')
    next_link=re.sub('page=\d','page='+str(page+1),url)
    
    for j in arti:
        
        link='http://www.todayenergy.kr/'+j.attrs['href']
        num+=1
        if num>20:
            num=num%20
        
        print('\n\n%d 페이지 %d 번째 기사입니다.\n\n' %(page,num))
        
        # 각 기사의 제목과 내용 스크래핑
        source=req.get(link).text
        arti_soup=bs(source,'html.parser')
        # 기사 제목
        title=arti_soup.select_one('#user-container > div.float-center.max-width-1280 > header > header > div').string
        print(title+'\n')
        # 기사 내용
        content=arti_soup.select('#article-view-content-div > p > span *')
        ######!!!!!!!!!!기사 내용 :10페이지부터는 또 dir이 달라짐...!!!!!
        # css 셀렉터 수준이 모두 달라 셀렉터를 *로 처리한 뒤 반복되는 구절 생략
        for k in range(1,len(content)):
            if content[k-1].text!=content[k].text:
                print(content[k-1].text.strip()+'\n')
        
        print('='*100)
        
    # 다음 페이지로 넘어감
    source=req.get(next_link).text
    soup=bs(source,'html.parser')
```
